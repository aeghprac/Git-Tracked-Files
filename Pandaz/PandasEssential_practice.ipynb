```python
### Which three countries have won most medals from 1984-2008?
# Make oo for recent years
recent = oo[oo.Edition >= 1984]
```

```python
# top 3 countries
recent.NOC.value_counts().head(3).plot(kind='bar',colormap= 'Pastel1')
```
