### Which US athlete has won the most medals in each edition of the olympics?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
oo =  pd.read_csv('data/olympics.csv', skiprows=4)
oo.head()
```

```python
usOnly = oo[oo.NOC == 'USA']
usOnly.head()
```

```python
usOnly = usOnly.groupby(['Edition', 'Athlete', 'Medal']).size().unstack('Medal', fill_value=0) # Just as I suspected, must assign here or doesn't work
usOnly['Total'] = (usOnly['Gold'] + usOnly['Silver'] + usOnly['Bronze'])
usOnly
```

```python
# ! This gets us part of the way there but we need to reset the index to move the 'Edition' index out to a column
usOnly.sort_values(['Edition','Total'], ascending=False)
```

```python
# Deleted a bunch of garbage here where I added too many fucking indexes and it turned into a disaster!
# Added way too many fucked up indexes, now we're good again
usOnly = usOnly.drop(columns=['level_0', 'index']) # Must assign this or do inplace 
usOnly
```

```python
# Now we're going to do a for loop through a weird nested groupby thing
# List to show the groupby object. It's composed of year and edition
list(usOnly.groupby('Edition'))
    
```

```python
# another way to show it
for year, group in usOnly.groupby('Edition'):
    print (year, group)
```

```python
for year, group in usOnly.groupby('Edition'):
    print(group.sort_values('Total', ascending=False)[:1]) # this is a genius move, just take the 0th item in the list
```

```python
# Now switch to a list comprehension
tu = [group.sort_values('Total', ascending=False)[:1] for year, group in usOnly.groupby('Edition')]
tu
```

```python
# Make an empty dataframe for the for loop to populate
top = pd.DataFrame()
for i in tu:
    top = top.append(i)
top
```
