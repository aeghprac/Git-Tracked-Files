Single line if statements are like this.  

```python
n = 666
print(str(n) + ': This is the number of the beast' if n > 665 and n < 667 else 'Nope') #This is called a ternary operator
# if n > 665 and n < 667:
#     print('The number of the beast')  #-This works too
```

```python
# 1, 2, Fizz, 4, Buzz, Fizz, 7, 8, Fizz, Buzz, 11, Fizz, 13, 14, FizzBuzz, 16
# Write an if-else statement that excludes those multiples

for n in range(101): #if you want to print the range between two numbers do this: range(1, 101)
    if n % 3 == 0:
        print('Fizz')
    elif n % 5 == 0:
        print('Buzz')
    elif n % 15 == 0:
        print('FizzBuzz')
    else:
        print(n) #this else statement at the end is optional
```
