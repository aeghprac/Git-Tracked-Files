# A simple program to find anagrams

```python
import pandas as pd
import numpy as np
```

```python
# Make an empty list and add all text from words text file to list
#
# Interestingly, the words_alpha.txt file is every single word in the English language
# that only includes alphabetic text (No symbols or numbers, etc.)

words = []
for line in open('words_alpha.txt', 'r'):
    words.append(line.strip().lower()) # remove whitespace and uppers
```

```python
# Convert to set to remove dupes
words = set(words)
```

```python
# Convert back to list and sort
words = sorted(words)
```

```python
# Now it's a sorted list with no duplicates. Let's check.
words
```

```python
# Define signature function (a list of all letters in a word)
def signature(word):
    signature = ''.join(sorted(word)) # Sort all letters of word and 'join' with blank character
    return signature
```

```python
# Example: this should return 'ehosu'
signature('house')
```

```python
# Define anagram finder -- returns all possible anagrams of a word as a list
def anagramFinder(myword):
    mysig = signature(myword)
    anagrams = []  # Defines a blank list
    
    for word in words: # all words from words.txt as defined above
        
        # Skip "self anagrams"
        if word == myword: 
            pass 
        
        # Add valid word to anagrams list
        elif mysig == signature(word):
            anagrams.append(word)
            
    if len(anagrams) == 0:
        print('Sorry, there were no anagrams for this word, dude.')
    else:
        return anagrams
    
```

```python
anagramFinder('crank')
```
